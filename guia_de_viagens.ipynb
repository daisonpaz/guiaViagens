```python
#Configurações iniciais
import google.generativeai as genai
from google.colab import userdata

GOOGLE_API_KEY=userdata.get('viagens')
try:
  genai.configure(api_key=GOOGLE_API_KEY)
except Exception as e:
  print(f"Erro ao configurar a API: {e}")
  exit()
```

```python
generation_config = {
  "candidate_count": 1,
  "temperature": 0.5,
}
```

```python
safety_settings={
    'HATE': 'BLOCK_NONE',
    'HARASSMENT': 'BLOCK_NONE',
    'SEXUAL' : 'BLOCK_NONE',
    'DANGEROUS' : 'BLOCK_NONE'
    }
```

```python
try:
  model = genai.GenerativeModel(model_name='gemini-1.0-pro',
                                  generation_config=generation_config,
                                  safety_settings=safety_settings,)
except Exception as e:
  print(f"Erro ao criar o modelo: {e}")
  exit()
```

```python
try:
  chat = model.start_chat(history=[])
except Exception as e:
  print(f"Erro ao iniciar o chat: {e}")
  exit()

pergunta1 = input('Qual sua idade?')
pergunta2 = input('Qual sua profissão?')
pergunta3 = input('Quais seus interesses?')
pergunta4 = input('Quanto você pode gastar? (em reais)')

prompt = f"Me dê opções de viagem, sabendo que tenho {pergunta1} de idade, que sou {pergunta2}, me interesso por {pergunta3} e posso investir até {pergunta4}."

while prompt != "fim":
  response = chat.send_message(prompt)
  print("Resposta:", response.text, '\n\n')
```

```python
#Melhorando a visualização
#Código disponível em https://ai.google.dev/tutorials/python_quickstart#import_packages
import textwrap
from IPython.display import display
from IPython.display import Markdown

def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

#Imprimindo o histórico
for message in chat.history:
  display(to_markdown(f'**{message.role}**: {message.parts[0].text}'))
  print('-------------------------------------------')
```
